```python
# # V-H (AC measurements) for different $\theta$
# # $H_{sw} - \theta$ 

# load Packages & Settings
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import os
# set fonts
font = {'family' : 'Sans',
        'weight' : 'bold',
        'size'   : 8}
rc('font', **font)

# get files from rootdir with specific suffix
def get_files(rootdir, suffix):
    file_list = []
    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            if file.endswith(suffix):
                file_list.append(os.path.join(subdir, file))

    file_list.sort(key=lambda x: x)
#     file_list.sort(key=lambda x: float(x.split('Index_')[1].split('Degree')[0]))
    return file_list

# load data from file and return it as numpy-array
def load_data(filename):
    arr = []
    for line in [line for line in open(filename, 'r')]:
        temp = [float(element) for element in line.split(',')] # parse each line and convert to floating
        temp =  np.array(temp[10:]) # dump the first M points
        temp.reshape(-1,1) # shape into columns
        arr.append(temp)
    del temp
    return np.array(arr)

# Normalize 1D data set to the range of [lo, hi]
def Normalize1D(x, lo, hi):
    # this works only for 1D data x
    return (hi - lo)*((x - min(x))/(max(x) - min(x)))-(hi - lo)/2.0

def search_jump(x, Angle):
    if(Angle <= 90):
        # find max for downward jump
        return np.argmin(x)
    elif(Angle > 90):
        # find min for upward jump
        return np.argmin(x)


# Move to root directory, load filenames and sort them by time.
rootdir = "/Users/yiyi/Downloads/Magnetism/Data_YIGPt_SW/Sample-01/YIGPt_RxxRxy_4kGTo-4kG_0To360Degree-20180107"
os.chdir(rootdir)

file_list = get_files(rootdir, '.txt')
# plot out V-H for different Angles
Angle_Hsw = []
i = 0
fig = plt.figure(figsize=(5, 5), dpi=300)
f, (ax1, ax2) = plt.subplots(2, sharex=True, sharey = False, figsize=(10, 16))
ax1.grid(False)
ax2.grid(False)
XTicks = np.linspace(-8, 8, 5)
YTicks_1 = np.linspace(-8, 8, 5)
YTicks_2 = np.linspace(-100, 100, 5)
# startIndex = 0
# endIndex = 

# startIndex = -175 #(Angle > 130 and Angle < 170)
# endIndex = -100
for file in file_list[0:1]:
    Angle_str = (file.split('Oe_')[1]).split('Degree')[0]
    Angle = round(float(Angle_str), 3) # get Angle in degree from filename
    data = load_data(file)
    data = data[:, :]
    data[1, :] = data[1, :]/1.0e3
    cumV = np.cumsum((Normalize1D(data[1, :], -1, 1)))
    if(Angle > 95 and Angle < 175):

#     if(Angle > 0 and Angle < 60):
#         index_jump = search_jump(cumV, Angle)
#         if ()
#         width =20
#         index_jump = search_jump(cumV[index_jump-width:index_jump+width], Angle) + index_jump - width
        if(i == 0):
            index_jump = search_jump(np.diff(data[1, :]), 90)
        else:
            index_jump = search_jump(np.diff(data[1, :]), 150)
        Hsw = data[0, index_jump]
        ax1.plot(data[0, :], data[1, :], marker='.', label=Angle_str) # reverse the axis to positive
#         ax1.plot(Hsw, data[1, index_jump], 'k+', markersize=20)
        ax1.set_xlim([-8, 8])
#         ax1.set_ylim([0, 20])
        ax1.set_xticks(XTicks)

        ax1.tick_params(labelsize=16)
#         ax1.axvline(x=-2.025, color='k')
        ax1.set_ylabel(r'$R_{xy} ( \Omega )$', fontsize=20)
#         ax2.plot(data[0, :], cumV, label=Angle_str)
        ax2.plot(data[0, :-1], np.diff(data[1, :]))
        ax2.set_xlim([-8, 8])
#         ax2.set_yticks(YTicks_1)
#         ax2.set_yticks(YTicks_2)
        ax2.tick_params(labelsize=16)
        ax2.set_ylabel(r'$\frac{d R_{xy}}{d H}$', fontsize=20)
        if((Hsw != data[0, -1:]) and (Hsw != data[0, 0])):
            Angle_Hsw.append([Angle, Hsw+2.025])
            ax1.plot(Hsw, data[1, index_jump], 'kx', markersize=20)
            ax2.plot(Hsw, np.diff(data[1, :])[index_jump], 'kx', markersize=20)
    i = i + 1
f.text(0.5, 0.9, '$H-R_{xy}$', ha='center', fontsize=28)
f.text(0.5, 0.04, '$H(Oe)$', ha='center', fontsize=20)
f.text(0.5, 0.04, '$H(Oe)$', ha='center', fontsize=20)
# f.text(0.45, 0.5, r'$H_{bias} = -2.025 Oe$', ha='center', fontsize=20)
# f.text(0.0, 0.5, '$V_{xy} (\mu V)$', va='center', rotation='vertical', fontsize=25)
f.savefig('PHE-90to180-1109To1110-Round_0.png')
np.savetxt('PHE-90to180-1109To1110-Round_0.csv', Angle_Hsw, delimiter=",")

# f.savefig('PHE-90to180-1109To1110-Round-' + rootdir.split('Round- ')[1] + '.png')
# np.savetxt('PHE-90to180-1109To1110-Round-' + rootdir.split('Round- ')[1] + '.csv', Angle_Hsw, delimiter=",")

# f.savefig(rootdir + '1107_PHE_1mA_73_Round-0.png')
# np.savetxt(rootdir + '1107_PHE_1mA_73_Round-0.csv', Angle_Hsw, delimiter=",")
# ax1.legend(loc='upper left', fontsize=10)
plt.show()
plt.close("all")
```
